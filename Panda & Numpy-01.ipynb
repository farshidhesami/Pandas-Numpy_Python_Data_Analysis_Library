### What is a DataFrame in pandas?

What is a DataFrame? A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

### DataFrame vs series in pandas?

Series can only contain single list with index, whereas dataframe can be made of more than one series or we can say that a dataframe is a collection of series that can be used to analyse the data.

### DataFrame - head() function

The head() function is used to get the first n rows. This function returns the first n rows for the object based on position. It is useful for quickly testing if your object has the right type of data in it. Number of rows to select.

Type/Default Value = int ==== Default Value: 5

Syntax: DataFrame.head(self, n=5)

https://www.w3resource.com/pandas/dataframe/dataframe-head.php#:~:text=DataFrame%20%2D%20head()%20function,type%20of%20data%20in%20it.&text=Number%20of%20rows%20to%20select.
